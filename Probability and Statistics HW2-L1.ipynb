Бросается игральная кость. Пусть событие A - появление чётного числа, событие B - появление числа больше трёх. Являются ли эти события независимыми?

```python
# событие А = 2,4,6
# событие B = 4,5,6
#изначально вероятности выпадения данных цифр у них одинаковые
A = 3/6 
B = 3/6
# но у A или B есть общие цифры: 4 и 6,
# поэтому при наступлении события A с цифрами (4 или 6) вероятность наступления события B будет снижаться и наоборот
A = 3/6 #выпало 4
B = 2/6 #теперь может выпасть только 5 и 6
#следовательно эти 2 события зависимые, так как одно событие может влиять на вероятность исхода другого
#если перефразировать, то условная вероятность события B при условии А не будет равна безусловной вероятности события B
#а это является подтверждением, что события зависими
```

Задача 6
Допустим, имеется некоторая очень редкая болезнь (поражает 0.1% населения). Вы приходите к врачу, вам делают тест на эту болезнь, и тест оказывается положительным. Врач говорит вам, что этот тест верно выявляет 99% больных этой болезнью и всего лишь в 1% случаев даёт ложный положительный ответ.
Вопрос: какова вероятность, что вы действительно больны ей?
Подсказка: вновь используйте формулу Байеса с раскрытием знаменателя с помощью формулы полной вероятности. 

```python
P1 = 0.001*0.99 #вероятность положительного теста с истинным диагнозом (болен, тест положительный)
P2 = 0.001*0.01 #вероятность положительного теста с ложным диагнозом (не болен, а тест положительный)
P3 = 0.999*0.99 #вероятность отрицательного теста с истинным диагнозом (не болен, тест отрицательный)
P4 = 0.999*0.01 #вероятность отрицательного теста с ложным диагнозом (болен, но тест отрицательный)

#проверим полную вероятность (должна быть = 1)
P = P1+P2+P3+P4
P
```
